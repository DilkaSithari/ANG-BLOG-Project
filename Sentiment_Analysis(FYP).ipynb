```python
import pandas as pd

from google.colab import drive
drive.mount("/content/drive", force_remount=True)

DATASET_PATH = '/content/drive/MyDrive/DataSets(FYP)/'
DATASET_1='EconomyNextAllNews.csv'

# Load your CSV file
df = pd.read_csv(DATASET_PATH + DATASET_1)
```

```python
# Check for missing values
print(df.isnull().sum())

# Drop rows with missing values (if any)
df = df.dropna()

# Alternatively, you can fill missing values with a placeholder
# df = df.fillna('N/A')
```

```python
# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Combine 'Headline' and 'News Description' into 'combined_text'
df['combined_text'] = df['Headline'] + ' ' + df['News Description']
```

```python
from transformers import pipeline

# Initialize the sentiment analysis pipeline
sentiment_pipeline = pipeline('sentiment-analysis')

# Function to get sentiment and score from combined text
def get_transformer_sentiment_score(text, neutral_threshold=0.5):
    result = sentiment_pipeline(text)[0]
    label = result['label'].lower()
    score = result['score']

    if score < neutral_threshold:
        return 'neutral', 0
    elif label == 'positive':
        return label, score
    elif label == 'negative':
        return label, -score
    else:
        return 'neutral', 0

# Apply sentiment analysis
df[['sentiment', 'sentiment_score']] = df['combined_text'].apply(lambda x: pd.Series(get_transformer_sentiment_score(x, neutral_threshold=0.5)))
```

```python
DATASET_2='EconomyNextAllNews_with_sentiments.csv'
# Save the updated DataFrame to a new CSV file
df.to_csv(DATASET_PATH + DATASET_2, index=False)

# Print some example rows to verify
print(df.head())
```
