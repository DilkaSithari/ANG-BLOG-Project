```python
import geopandas
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import files
dataload = files.upload()
```

```python
df= pd.read_csv('Data_SL.csv',index_col=0)
df
```

```python
df['geometry']= geopandas.GeoSeries.from_wkt(df['geometry'].astype('str'))
gdf = geopandas.GeoDataFrame(df,crs="EPSG: 4326")
gdf
```

Display latitudinal and longitudinal values, x- latitude values, y- logitude values

```python
gdf.plot()
```

```python
gdf_mer = gdf.to_crs(epsg = 3857)
```

converting wgs84 to web mercarator/ sudo mercarator, good for visualization, y- web mercarator y, x- web mercarator x

```python
gdf_mer.plot()
```

```python
fig, ax=plt.subplots(1,2,figsize=(10,5))
gdf.plot(ax=ax[0],color='red')
ax[0].set_title('WGS 84')
ax[0].set_xlabel('Longitude')
ax[0].set_ylabel('Latitude')

gdf_mer.plot(ax=ax[1],color='blue')
ax[1].set_title('New CRS')
ax[1].set_xlabel('Web Mercator X')
ax[1].set_ylabel('Web Mercator Y')

plt.tight_layout()
plt.show()
```

```python
temperature= [39,24,32,37,28,35,33,37,35]
gdf['temperature']= temperature
gdf
```

```python
rain=[24,33,43,20,60,40,10,25,38]
gdf['rain']= rain
gdf
```

```python
gdf['pop_den']=gdf['population']/gdf['area']
gdf
```

```python
gdf.plot(column='pop_den',legend= True, cmap='plasma',legend_kwds={'label': "Population Density of Sri Lanka"})
```

```python
gdf.plot(column='rain',legend= True, cmap='plasma',legend_kwds={'label': "Rainfall"})
```
